## 四分位範囲

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec

# ===== 入力データ =====
group0_row_data = {
    "Academic citation 2011": treat_data['academic_citation_2011'].tolist(),
    "Academic citation 2012": treat_data['academic_citation_2012'].tolist(),
    "Academic citation 2013": treat_data['academic_citation_2013'].tolist(),
    "Academic citation 2014": treat_data['academic_citation_2014'].tolist(),
    "Academic citation 2015": treat_data['academic_citation_2015'].tolist(),
    "IGO policy citation 2015": treat_data['policy_citation_2015'].tolist(),
    "Publication numbers as of 2015": treat_data['paper_num_2015'].tolist()
}

group1_row_data = {
    "Academic citation 2011": control_data['academic_citation_2011'].tolist(),
    "Academic citation 2012": control_data['academic_citation_2012'].tolist(),
    "Academic citation 2013": control_data['academic_citation_2013'].tolist(),
    "Academic citation 2014": control_data['academic_citation_2014'].tolist(),
    "Academic citation 2015": control_data['academic_citation_2015'].tolist(),
    "IGO policy citation 2015": control_data['policy_citation_2015'].tolist(),
    "Publication numbers as of 2015": control_data['paper_num_2015'].tolist()
}

# グループラベル
group0_label = "Treatment(n=101)"  # オレンジ
group1_label = "Control(n=101)"    # 青

manual_x_ranges = {
    "Academic citation 2011": (0, 6000),
    "Academic citation 2012": (0, 6000),
    "Academic citation 2013": (0, 6000),
    "Academic citation 2014": (0, 6000),
    "Academic citation 2015": (0, 6000),
    "IGO policy citation 2015": (0, 100),
    "Publication numbers as of 2015": (0, 200)
}

# ===== 中央値と四分位範囲を算出 =====
metrics = list(group0_row_data.keys())
group0_stats = {}
group1_stats = {}

for metric in metrics:
    # Group 0
    data0 = np.array(group0_row_data[metric])
    median0 = np.median(data0)
    q1_0 = np.percentile(data0, 25)
    q3_0 = np.percentile(data0, 75)
    group0_stats[metric] = [median0, q1_0, q3_0]

    # Group 1
    data1 = np.array(group1_row_data[metric])
    median1 = np.median(data1)
    q1_1 = np.percentile(data1, 25)
    q3_1 = np.percentile(data1, 75)
    group1_stats[metric] = [median1, q1_1, q3_1]

# ===== プロット設定 =====
fig = plt.figure(figsize=(14, 16))
gs = GridSpec(len(metrics), 1, figure=fig)

# 凡例用ダミー
fig.add_subplot(111).set_visible(False)
plt.plot([], [], 'o-', color='#0000FF', label=group0_label)
plt.plot([], [], 'o-', color='#FF0000', label=group1_label)
plt.legend(loc='upper center', bbox_to_anchor=(0.5, 1.05), ncol=2, frameon=False, fontsize=12)

for i, metric in enumerate(metrics):
    ax = fig.add_subplot(gs[i])

    median0, q1_0, q3_0 = group0_stats[metric]
    median1, q1_1, q3_1 = group1_stats[metric]

    # Group 1 (青)
    ax.plot(median1, 0, 'o', color='#FF0000', markersize=12)
    ax.hlines(0, q1_1, q3_1, color='#FF0000', linewidth=3)
    ax.vlines([q1_1, q3_1], -0.1, 0.1, color='#FF0000', linewidth=3)

    # Group 0 (オレンジ)
    ax.plot(median0, 0, 'o', color='#0000FF', markersize=12)
    ax.hlines(0, q1_0, q3_0, color='#0000FF', linewidth=3)
    ax.vlines([q1_0, q3_0], -0.1, 0.1, color='#0000FF', linewidth=3)

    # x軸の範囲設定
    if metric in manual_x_ranges:
        x_min, x_max = manual_x_ranges[metric]
    else:
        combined = np.concatenate([data0, data1])
        margin = (combined.max() - combined.min()) * 0.1
        x_min = combined.min() - margin
        x_max = combined.max() + margin

    ax.set_xlim(x_min, x_max)
    ticks = np.linspace(x_min, x_max, 5)
    ax.set_xticks(ticks)
    ax.set_xticklabels([f"{tick:.1f}" if x_max<100 else f"{int(tick)}" for tick in ticks], fontsize=25)

    ax.set_yticks([])
    ax.set_ylim(-0.5, 0.5)
    ax.set_ylabel(metric, fontsize=20, rotation=0, ha='right', va='center', labelpad=15)
    ax.grid(axis='x', linestyle='-', alpha=0.2)
    for spine in ['top', 'right', 'left']:
        ax.spines[spine].set_visible(False)

plt.tight_layout(rect=[0, 0, 0.95, 1])
plt.subplots_adjust(hspace=0.5)
plt.show()
```
