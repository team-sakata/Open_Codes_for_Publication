```python
import numpy as np
from scipy.stats import brunnermunzel

# サンプルデータ
group_A = treat_paper_df_tmp0_d['D_score'].tolist()
group_B = treat_paper_df_tmp1_d['D_score'].tolist()
group_C = treat_paper_df_tmp2_d['D_score'].tolist()


# ブルンナー＝ムンツェル検定の実行
statistic, p_value = brunnermunzel(group_C, group_A, alternative='greater')

print("Brunner-Munzel 検定統計量:", statistic)
print("p値:", p_value)
print("p値:", p_value<0.0167)
```

```python
# Cliff's Delta を計算する関数
def cliffs_delta(x, y):
    n_x = len(x)
    n_y = len(y)
    # 各要素のペアの差分を求める（外積的な計算）
    diff = np.subtract.outer(x, y)  # shape: (n_x, n_y)
    # x > y の場合は +1、x < y の場合は -1、等しい場合は 0 とする
    n_pos = np.sum(diff > 0)
    n_neg = np.sum(diff < 0)
    delta = (n_pos - n_neg) / (n_x * n_y)
    return delta

# 計算実行
# サンプルデータ
treat_data = treat_paper_df_tmp2_d['D_score'].tolist()  
control_data = treat_paper_df_tmp0_d['D_score'].tolist()    


delta = cliffs_delta(treat_data, control_data)
print("Cliff's Delta =", delta)
```
