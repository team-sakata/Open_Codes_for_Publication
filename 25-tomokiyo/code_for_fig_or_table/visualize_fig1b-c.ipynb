```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import ttest_ind, sem


# Define the years for analysis
years = [f'academic_citation_{year}' for year in range(2011, 2024)]
relevant_data = matched_data[['guideline_member'] + years]


# Apply log10 transformation (log10(x + 1)) to citation data
for year in years:
    relevant_data[year] = np.log10(relevant_data[year] + 1)

# Initialize lists to store means, confidence intervals, and p-values
log_academic_means_group_0 = []
log_academic_means_group_1 = []
log_academic_conf_ints_group_0 = []
log_academic_conf_ints_group_1 = []
log_academic_errors_group_0 = []
log_academic_errors_group_1 = []
log_academic_p_values = []

# Confidence level
conf_level = 0.95
z_value = 1.96  # for 95% confidence level

# Iterate over years to calculate statistics
for year in years:
    group_0 = relevant_data[relevant_data['guideline_member'] == 0][year]
    group_1 = relevant_data[relevant_data['guideline_member'] == 1][year]

    # Calculate means
    mean_0 = group_0.mean()
    mean_1 = group_1.mean()
    log_academic_means_group_0.append(mean_0)
    log_academic_means_group_1.append(mean_1)

    log_academic_errors_group_0.append(group_0.sem())
    log_academic_errors_group_1.append(group_1.sem())

    # Perform t-test
    _, p_value = ttest_ind(group_0, group_1, nan_policy='omit', equal_var=False, alternative='less')
    log_academic_p_values.append(p_value)

    # Leveneの等分散検定
    statistic, p_value = stats.levene(group_0, group_1)
    
    print(f"検定統計量: {statistic}")
    print(f"p値: {p_value}")
    
    # 有意水準を設定
    alpha = 0.05

    print(year)
    
    if p_value > alpha:
        print("p値が有意水準を上回るため、2つの配列の分散は等しいと判断されます。")
    else:
        print("p値が有意水準以下のため、2つの配列の分散は等しくないと判断されます。")
        print()



academic_means_group_0 = []
academic_means_group_1 = []
academic_conf_ints_group_0 = []
academic_conf_ints_group_1 = []
academic_errors_group_0 = []
academic_errors_group_1 = []
academic_p_values = []
academic_group_0_Q1 = []
academic_group_0_Q3 = []
academic_group_1_Q1 = []
academic_group_1_Q3 = []

for year in years:
    group_0 = matched_data[matched_data['guideline_member'] == 0][year]
    group_1 = matched_data[matched_data['guideline_member'] == 1][year]


    # Calculate means
    mean_0 = group_0.median()
    mean_1 = group_1.median()
    academic_means_group_0.append(mean_0)
    academic_means_group_1.append(mean_1)

    #academic_errors_group_0.append(group_0.sem())
    #academic_errors_group_1.append(group_1.sem())


    # 四分位のエラーバーの場合
    
    group_0_Q1 = group_0.quantile(0.25)
    group_0_Q3 = group_0.quantile(0.75)
    # 四分位偏差を算出
    group_0_QD = (group_0_Q3 - group_0_Q1) / 2
    academic_group_0_Q1.append(group_0_Q1)
    academic_group_0_Q3.append(group_0_Q3)


    group_1_Q1 = group_1.quantile(0.25)
    group_1_Q3 = group_1.quantile(0.75)
    # 四分位偏差を算出
    group_1_QD = (group_1_Q3 - group_1_Q1) / 2
    academic_group_1_Q1.append(group_1_Q1)
    academic_group_1_Q3.append(group_1_Q3)


    academic_errors_group_1.append(group_1_QD)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import ttest_ind, sem


# Define the years for analysis
years = [f'policy_citation_{year}' for year in range(2011, 2024)]
relevant_data = matched_data[['guideline_member'] + years]


# Apply log10 transformation (log10(x + 1)) to citation data
for year in years:
    relevant_data[year] = np.log10(relevant_data[year] + 1)

# Initialize lists to store means, confidence intervals, and p-values
log_policy_means_group_0 = []
log_policy_means_group_1 = []
log_policy_conf_ints_group_0 = []
log_policy_conf_ints_group_1 = []
log_policy_errors_group_0 = []
log_policy_errors_group_1 = []
log_policy_p_values = []

# Confidence level
conf_level = 0.95
z_value = 1.96  # for 95% confidence level

# Iterate over years to calculate statistics
for year in years:
    group_0 = relevant_data[relevant_data['guideline_member'] == 0][year]
    group_1 = relevant_data[relevant_data['guideline_member'] == 1][year]

    # Calculate means→中央値に変更
    mean_0 = group_0.mean()
    mean_1 = group_1.mean()
    log_policy_means_group_0.append(mean_0)
    log_policy_means_group_1.append(mean_1)

    log_policy_errors_group_0.append(group_0.sem())
    log_policy_errors_group_1.append(group_1.sem())

    # Perform t-test
    # _, p_value = ttest_ind(group_0, group_1, nan_policy='omit', equal_var=False, alternative='less')
    mannwhitney_result = stats.mannwhitneyu(group_0, group_1, alternative='less')
    p_value = mannwhitney_result.pvalue
    log_policy_p_values.append(p_value)

    # Leveneの等分散検定
    statistic, p_value = stats.levene(group_0, group_1)
    
    print(f"検定統計量: {statistic}")
    print(f"p値: {p_value}")
    
    # 有意水準を設定
    alpha = 0.05

    print(year)
    
    if p_value > alpha:
        print("p値が有意水準を上回るため、2つの配列の分散は等しいと判断されます。")
    else:
        print("p値が有意水準以下のため、2つの配列の分散は等しくないと判断されます。")
        print()

policy_means_group_0 = []
policy_means_group_1 = []
policy_conf_ints_group_0 = []
policy_conf_ints_group_1 = []
policy_errors_group_0 = []
policy_errors_group_1 = []
policy_p_values = []
policy_group_0_Q1 = []
policy_group_0_Q3 = []
policy_group_1_Q1 = []
policy_group_1_Q3 = []

for year in years:
    group_0 = matched_data[matched_data['guideline_member'] == 0][year]
    group_1 = matched_data[matched_data['guideline_member'] == 1][year]

    # Calculate means
    mean_0 = group_0.median()
    mean_1 = group_1.median()
    policy_means_group_0.append(mean_0)
    policy_means_group_1.append(mean_1)

    #policy_errors_group_0.append(group_0.sem())
    #policy_errors_group_1.append(group_1.sem())
    

    group_0_Q1 = group_0.quantile(0.25)
    group_0_Q3 = group_0.quantile(0.75)
    # 四分位偏差を算出
    group_0_QD = (group_0_Q3 - group_0_Q1) / 2

    group_1_Q1 = group_1.quantile(0.25)
    group_1_Q3 = group_1.quantile(0.75)
    # 四分位偏差を算出
    group_1_QD = (group_1_Q3 - group_1_Q1) / 2

    
    policy_group_0_Q1.append(group_0_Q1)
    policy_group_0_Q3.append(group_0_Q3)


    policy_group_1_Q1.append(group_1_Q1)
    policy_group_1_Q3.append(group_1_Q3)

    


```
