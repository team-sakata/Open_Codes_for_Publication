```python
import networkx as nx
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

import math


def network_pos(G):
    
    # 
    largest_cc = max(nx.connected_components(G), key=len)
    subgraph = G.subgraph(largest_cc)
    
    # Kamada-Kawai
    # pos = nx.kamada_kawai_layout(subgraph)
    pos = kamada(subgraph)
    #pos = nx.spring_layout(subgraph)
    #pos = nx.forceatlas2_layout(subgraph)
    
    #  (: )
    
    # isolated_nodes = [n for n in G.nodes() if nx.is_isolate(G, n)]
    isolated_nodes = [n for n in G.nodes() if n not in subgraph.nodes()]
    
    num_isolated = len(isolated_nodes)
    radius = 6.5  # 
    for i, node in enumerate(isolated_nodes):
        angle = 2 * math.pi * i / num_isolated  # 
        x = radius * math.cos(angle)  # x
        y = radius * math.sin(angle)  # y
        pos[node] = (x, y)
        
    node_colors = []
    for c in G.nodes():
        if c in authid_12:
            node_colors.append('blue')
        elif c in authid_13:
            node_colors.append('green')
        elif c in authid_14:
            node_colors.append('brown')
        elif c in authid_15:
            node_colors.append('red')
        else:
            print(c)
            node_colors.append('gray')

    return pos, node_colors
```
