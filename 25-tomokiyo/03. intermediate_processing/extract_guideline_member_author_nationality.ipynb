```python
# Define the region mapping dictionary
region_mapping = {
    'North_America': ['usa', 'can', 'mex'],
    'East_Asia': ['chn', 'jpn', 'kor', 'twn', 'hkg', 'mac'],
    'Europe': ['gbr', 'fra', 'deu', 'ita', 'esp', 'nld', 'bel', 'lux', 'swe', 'dnk', 'fin', 'nor', 'che', 'aut', 'irl', 'prt', 'grc', 'pol', 'hun', 'cze', 'svk', 'svn', 'hrv', 'rou', 'bgr', 'est', 'lva', 'ltu', 'rus', 'ukr', 'blr'],
    'South_America': ['bra', 'arg', 'chl', 'per', 'col', 'ury', 'pry', 'ecu', 'bol', 'ven', 'guy', 'sur'],
    'Africa': ['nga', 'zaf', 'egy', 'eth', 'ken', 'mar', 'dza', 'tza', 'ago', 'zmb', 'uga', 'gha', 'mwi', 'sen', 'tun', 'civ', 'mlt', 'cmr', 'moz', 'bfa', 'mli', 'tgo', 'sdn', 'gmb', 'rwa', 'sle', 'ben', 'caf', 'gnb', 'lso', 'gnq', 'com', 'swz', 'eri', 'stp', 'lby', 'cpv', 'mdg', 'syc'],
    'Middle_East': ['sau', 'are', 'qat', 'kwt', 'bhr', 'omn', 'jor', 'irq', 'irn', 'lbn', 'syr', 'pse', 'isr'],
    'Oceania': ['aus', 'nzl', 'png', 'fji', 'wsm', 'ton', 'vut', 'plw', 'tto'],
    'South_Asia': ['ind', 'pak', 'bgd', 'npl', 'lka', 'afg', 'mdv', 'btn'],
    'Southeast_Asia': ['mys', 'sgp', 'phl', 'idn', 'tha', 'vnm', 'khm', 'lao', 'brn', 'mmr', 'tls'],
    'Central_America': ['pan', 'cri', 'hnd', 'nic', 'slv', 'gtm', 'blz'],
    'Caribbean': ['cub', 'jam', 'hti', 'dom', 'bhs', 'brb', 'tto', 'dma', 'vct', 'kna', 'lca', 'grd', 'atg', 'msr', 'vgb', 'cym', 'pri', 'vir', 'ant'],
    'Central_Asia': ['kaz', 'kgz', 'tjk', 'tkm', 'uzb', 'arm', 'aze', 'geo'],
    'Other': ['asm', 'ata', 'fro', 'guf', 'gib', 'glp', 'gum', 'pyf', 'reu', 'spm', 'wlf']
}

# Define a dictionary to convert country names to ISO 3-letter codes
country_to_iso = {
    'Argentina': 'arg', 'Australia': 'aus', 'Azerbaijan': 'aze', 'Belgium': 'bel',
    'Brazil': 'bra', 'Canada': 'can', 'China': 'chn', 'Finland': 'fin', 'France': 'fra',
    'Germany': 'deu', 'Ghana': 'gha', 'India': 'ind', 'Indonesia': 'idn', 'Iran': 'irn',
    'Israel': 'isr', 'Italy': 'ita', 'Japan': 'jpn', 'Korea': 'kor', 'Kuwait': 'kwt',
    'Lebanon': 'lbn', 'Mexico': 'mex', 'Nepal': 'npl', 'Netherlands': 'nld',
    'Philippines': 'phl', 'Russia': 'rus', 'Senegal': 'sen', 'South Africa': 'zaf',
    'Spain': 'esp', 'Switzerland': 'che', 'Thailand': 'tha', 'United Kingdom': 'gbr',
    'United States': 'usa'
}

# List of countries to classify
countries = [
    'Argentina', 'Australia', 'Azerbaijan', 'Belgium', 'Brazil', 'Canada', 'China', 
    'Finland', 'France', 'Germany', 'Ghana', 'India', 'Indonesia', 'Iran', 'Israel', 
    'Italy', 'Japan', 'Korea', 'Kuwait', 'Lebanon', 'Mexico', 'Nepal', 'Netherlands', 
    'Philippines', 'Russia', 'Senegal', 'South Africa', 'Spain', 'Switzerland', 
    'Thailand', 'United Kingdom', 'United States'
]

# Create a new dictionary to classify countries by region
classified_countries = {}

for country in countries:
    iso_code = country_to_iso[country]
    for region, iso_list in region_mapping.items():
        if iso_code in iso_list:
            classified_countries[country] = region
            break

import pandas as pd
classified_countries_df = pd.DataFrame.from_dict(classified_countries, orient="index", columns=["region"])

region_order = ['North_America', 'Caribbean', 
                'Central_America', 'South_America', 'Europe', 'East_Asia', 
                'South_Asia', 'Southeast_Asia', 'Central_Asia', 
                'Middle_East', 'Africa', 'Oceania', 'Other']

# Set up unique colors for each region
color_map = cm.get_cmap('tab20', len(region_order))  # Use 'tab20' to get distinct colors
colors = [color_map(i) for i in range(len(region_order))]

# Set up a color dictionary based on region order
color_dict = {region: colors[i] for i, region in enumerate(region_order)}
```

```python

# Create a new column 'region' in the dataframe based on the mapping
def assign_region(country):
    for region, countries in region_mapping.items():
        if country in countries:
            return region
    return 'Other'

# classified_countries_df['region'] = classified_countries_df['country'].apply(assign_region)
classified_countries_df['color'] = classified_countries_df['region'].map(color_dict)
```

```python
classified_countries_df.reset_index(inplace=True)
```

```python
classified_countries_df.rename(columns={'index':'country'},inplace=True)
```
